```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
from sklearn.metrics import r2_score, mean_squared_error

from sklearn_pandas import DataFrameMapper

import scattertext as st
from pprint import pprint
```

```python
# read in the data

data = pd.read_csv("winemag-data-130k-v2.csv")
```

# Which words are used to describe good wine and bad wine?

```python
# create a new dataframe with a column where wines are divided in good and bad by 90 points

def points_to_category(points):
    '''classfy point rating into good/bad category'''
    if points < 90:
        return 'bad'
    else:
        return 'good'
clean_data_cat = clean_data.copy()
clean_data_cat['points_cat'] = clean_data_cat.points.apply(points_to_category)
clean_data_cat.points_cat.head()
```

```python
# create an interactive html based graphic where the frequency of words in description for good and bad wines is illustrated
# code and package for this graph can be found here https://github.com/JasonKessler/scattertext


from scattertext import SampleCorpora, whitespace_nlp_with_sentences
from scattertext import produce_scattertext_explorer
from scattertext.CorpusFromPandas import CorpusFromPandas
from scattertext.termscoring.ScaledFScore import ScaledFScorePresets

convention_df = clean_data_cat
corpus = CorpusFromPandas(convention_df,
                          category_col='points_cat',
                          text_col='description',
                          nlp=whitespace_nlp_with_sentences).build()

html = produce_scattertext_explorer(
	corpus,
	category='good',
	category_name='good wine',
	not_category_name='bad wine',
	minimum_term_frequency=10,
	pmi_threshold_coefficient=8,
	width_in_pixels=1200,
)

open('./winescatter.html', 'wb').write(html.encode('utf-8'))
print('Open ./winescatter.html in your Browser.')
```

# Which country should you buy your favourite variety from?

# Are old wines better and more expensive?
<br>
<br>
In Oenology, the science and study of wine, there is not only one definition of what old is, so I show the borderline
between old and newer wine for different age limits.

```python
# create new dataframe

wine_year = clean_data.copy()

# extract the year out of the 'title' column

wine_year['year'] = wine_year['title'].str.extract('([^0-9][0-9]{4}[^0-9])')[0]

# delete all data which is not plausible or relevant (all rows with no value and older than 1700 and newer than 2018)

wine_year['year'].fillna(2100, inplace=True)
wine_year['year'] = wine_year['year'].astype(int)
wine_year = wine_year[(wine_year['year'] < 2018 ) & (wine_year['year'] > 1700)]

# wines of before 1938 (80 years old) are not cheap, so the year is probably wrong if they are under 100 $

wine_year = wine_year[~((wine_year['year'] < 1938 ) & (wine_year['price'] < 100))]

# calculate an age column

wine_year['age'] = 2018 - wine_year['year']
```

```python
# plot age against points and against price with an example old_limit (borderline for old wine) at 25 years

fig, ax = plt.subplots()
fig.set_size_inches(16,10)

plt.subplot(1, 2, 1)
plt.scatter(wine_year['age'], wine_year['points'])
plt.title('Age of wine vs. review points')
plt.xlabel('age')
plt.ylabel('points')
plt.axvline(x=25, c='r', label='old_limit = 25')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(wine_year['age'], wine_year['price'])
plt.title('Age of wine vs. price')
plt.xlabel('age')
plt.ylabel('price')
plt.axvline(x=25, c='r', label='old_limit = 25')

plt.legend()
plt.show()
```

```python
# calculate average prices and points for different old_limits (age when wine is considered to be old)

avg_price_old_list, avg_price_young_list, avg_points_old_list, avg_points_young_list = list(), list(), list(), list()

for year in range(20, 51, 2):

    # calculate average price of old wines

    avg_price_old = wine_year[wine_year['age'] > year]['price'].mean()
    avg_price_old_list.append(avg_price_old)
    
    # calculate average price of non old wines

    avg_price_young = wine_year[wine_year['age'] <= year]['price'].mean()
    avg_price_young_list.append(avg_price_young)
    
    # calculate average points of old wines

    avg_points_old = wine_year[wine_year['age'] > year]['points'].mean()
    avg_points_old_list.append(avg_points_old)
    
    # calculate average points of non old wines

    avg_points_old = wine_year[wine_year['age'] <= year]['points'].mean()
    avg_points_young_list.append(avg_points_old)
```

```python
# building dataframe out of values from different old_limits

# create old_limit column_values

index_list = list()
for x in range(20, 51, 2):
    index_list.append(x)
index_list

# create dictionary as input for dataframe

d = {'avg_price_old': avg_price_old_list, 'avg_price_young': avg_price_young_list, 'avg_points_old': avg_points_old_list,
     'avg_points_young': avg_points_young_list, 'old_limit':index_list, 'diff_price': np.array(avg_price_old_list)
     - np.array(avg_price_young_list), 'diff_points': np.array(avg_points_old_list) - np.array(avg_points_young_list) }

#create and fill the dataframe

old_wine_df = pd.DataFrame(d)
old_wine_df
```

```python
# plot development of average price and points for different old_limits

fig, ax = plt.subplots()
fig.set_size_inches(16,5)

fig.suptitle('Difference caused by definitions of "old"', fontsize=16)

plt.subplot(1, 2, 1)
plt.plot(old_wine_df['old_limit'], old_wine_df['avg_price_old'], label='avg_price_old')
plt.plot(old_wine_df['old_limit'], old_wine_df['avg_price_young'], c='r', label='avg_price_young')
plt.plot(old_wine_df['old_limit'], old_wine_df['diff_price'], c='g', label='diff_price')

plt.title('differences by price')
plt.xlabel('agelimit')
plt.ylabel('avergae price')

plt.legend()

plt.subplot(1, 2, 2)
plt.plot(old_wine_df['old_limit'], old_wine_df['avg_points_old'], label='avg_points_old')
plt.plot(old_wine_df['old_limit'], old_wine_df['avg_points_young'], c='r', label='avg_points_young')
plt.plot(old_wine_df['old_limit'], old_wine_df['diff_points'], c='g', label='diff_price')

plt.title('differences by points')
plt.xlabel('age limit')
plt.ylabel('average points')

plt.legend()

plt.show()
```

In the figure above you can see that the difference in prices is rising pretty fast and the difference in points rises slow when you increase the number of years which determine the definition of old. The small increase of the price and points for young is due to the bigger number of reviews of newer wines, so if you add some old wines the average will not increase much. But you can see the trend that old wines become much more expensive with age and only a little bit better.

# Would an algorithm give the same rating?

```python
# use wine_year because it is already cleaned and has the additional feature year/age

wine_year.info()
```

```python
# delete 'Unnamed : 0' and 'year' because 'Unnamed : 0' has index like information and 'year' information is stored in age

wine_ml = wine_year.drop( ['year', 'Unnamed: 0'], axis=1)
```

```python
# check for duplicates in the description column

wine_ml['description'].duplicated().sum()
```

```python
# remove all duplicates

wine_ml = wine_ml[~wine_ml['description'].duplicated()]
```

```python
# analyze the number of unique values in the categorical columns for one-hot-encoding

non_numeric_columns = wine_ml.dtypes[wine_ml.dtypes=='object'].index.tolist()
num_unique_values = []

for column in non_numeric_columns:
    num_unique_values.append(len(data[column].value_counts()))
    
pd.Series(num_unique_values, non_numeric_columns)
```

```python
# 'title' and 'variety' have too much unique values for rational one-hot-encoding, 'description' will be countvectorized

wine_ml.drop(['title', 'winery'], axis=1, inplace=True)
```

```python
# dataframe has now 4 non-numeric categorical columns

wine_ml.head(3)
```

```python
# encode labels for unique values in the categorical columns

cat_columns = ['country', 'province', 'variety']

country_dummies = pd.get_dummies(wine_ml['country'])
province_dummies = pd.get_dummies(wine_ml['province'])
variety_dummies = pd.get_dummies(wine_ml['variety'])

# drop columns which are encoded

wine_ml.drop(cat_columns, axis=1, inplace=True)

# add one-hot-code columns to the dataframe

wine_ml = pd.concat([wine_ml, country_dummies, variety_dummies], axis=1)
```

```python
# show one-hot-encoded dataframe

wine_ml.head(3)
```

```python
# Usage of DataFrameMapper from sklearn_pandas to use 'description' as text and other columns for prediction

# create mapperlist for all columns to fill the DataFrameMapper

mapper_list = []

columns_without_descritpion = wine_ml.columns.tolist()
columns_without_descritpion.remove('description')
columns_without_descritpion.remove('points')



for column in columns_without_descritpion:
    mapper_list.append((column, None))

# use CountVectorizer with elimination of stopwords

mapper_list.append(('description', CountVectorizer(stop_words = 'english')))
mapper_list
```

```python
# instantiate DataFrameMapper

mapper = DataFrameMapper(mapper_list)
```

```python
# specify features and the categories to predict, due to limitation of my machine 10,000 random values will fit the memory

X = mapper.fit_transform(wine_ml.sample(n=10000, random_state=123))
y = wine_ml.points.sample(n=10000, random_state=123)
```

```python
# split data into train and test units 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
# fit AdaBoostRegressor

clf1 = AdaBoostRegressor(random_state=123)
clf1.fit(X_train, y_train)
```

```python
# fit GradientBoostRegressor

clf2 = GradientBoostingRegressor(random_state=123)
clf2.fit(X_train, y_train)
```

```python
# predict categories and show the score for AdaBoostRegressor

predictions = clf1.predict(X_test)
print('AdaBoostRegressor')
print('r2_score: {} mean_squarred_error: {}'.format(r2_score(y_test, predictions), mean_squared_error(y_test, predictions)))

# predict categories and show the score for GradientBoostRegressor

predictions2 = clf2.predict(X_test)
print('GradientBoostRegressor')
print('r2_score: {} mean_squarred_error: {}'.format(r2_score(y_test, predictions2), mean_squared_error(y_test, predictions2)))
```

```python
# plot the deviations of the predictions from y_test

fig, ax = plt.subplots()
fig.set_size_inches(5,12)

plt.subplot()
plt.title('Prediction-Error-Boxplot')
plt.ylabel('Difference in points')
plt.boxplot(predictions2-y_test, range(len(predictions2-y_test)))
plt.show()
```

The idea of this machine learning model was to combine the vectorized textdata('description'), the numerical data('price', 'age') and categorical data('country', 'province', 'variety') in order to give a point rating like a human being would give. The result is not very good. r2_score and mean_squarred_error are slightly better for the GradientBostRegressor. The result could be optimized if all of the data and hyperparameter tuning would be used on a more powerful machine(currently only 10,000 out of 120,000 are used). The downside of this algorithm is, that you still need a human to get a 'description' of a wine. The other data could be scraped out of other sources, so it is still dependend on somebody to interact with the wine to make the machine to predict a point rating.
